```python
import numpy as np
import pandas as pd
from sklearn import datasets
```

```python
house_data = datasets.load_boston()
```

```python
df = pd.DataFrame(house_data.data, columns = house_data.feature_names)
```

```python
df.head()
```

```python
df['MEDV'] = house_data.target
```

```python
df.info()
```

```python
X = df['RM'].values
y = df['MEDV'].values
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
plt.xlabel('RM')
plt.ylabel('MEDV')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = X.reshape(-1,1)
lr = LinearRegression()
lr.fit(X,y)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X))
plt.xlabel('RM')
plt.ylabel('MEDV')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
y_pred = lr.predict(X)
mse = mean_squared_error(y,y_pred)
r2 = r2_score(y,y_pred)
print("Linear Regression MSE is: %2.2f and R2 is: %2.2f" % (mse,r2))
```

```python
from sklearn.linear_model import Ridge, Lasso, ElasticNet
ridge = Ridge(alpha=1.0)
lasso = Lasso(alpha=1.0)
elanet = ElasticNet(alpha=1,l1_ratio=0.5)
ridge.fit(X,y)
lasso.fit(X,y)
elanet.fit(X,y)
y1_pred = ridge.predict(X)
y2_pred = lasso.predict(X)
y3_pred = elanet.predict(X)
plt.scatter(X,y)
plt.plot(X,y_pred,label='Linear Regression')
plt.plot(X,y1_pred,label='Ridge')
plt.plot(X,y2_pred,label='Lasso')
plt.plot(X,y3_pred,label='ElasticNet')
plt.xlabel('RM')
plt.ylabel('MEDV')
plt.legend()
plt.show()
```

```python
mse1 = mean_squared_error(y,y1_pred)
r21 = r2_score(y,y1_pred)
mse2 = mean_squared_error(y,y2_pred)
r22 = r2_score(y,y2_pred)
mse3 = mean_squared_error(y,y3_pred)
r23 = r2_score(y,y3_pred)
print("Ridge MSE is: %2.2f and R2 is: %2.2f" % (mse1,r21))
print("Lasso MSE is: %2.2f and R2 is: %2.2f" % (mse2,r22))
print("ElasticNet MSE is: %2.2f and R2 is: %2.2f" % (mse3,r23))
```

```python
from sklearn.linear_model import RANSACRegressor
ransac = RANSACRegressor(LinearRegression(),max_trials=100,min_samples=50,loss='absolute_loss',residual_threshold=0.5,random_state=1)
ransac.fit(X,y)
y4_pred = ransac.predict(X)
plt.scatter(X,y)
plt.plot(X,y4_pred,label='RANSAC Regression')
plt.xlabel('RM')
plt.ylabel('MEDV')
plt.legend()
plt.show()
```

```python
mse4 = mean_squared_error(y,y4_pred)
r24 = r2_score(y,y4_pred)
print("RANSAC MSE is: %2.2f and R2 is: %2.2f" % (mse4,r24))
```

```python
X = df['LSTAT'].values
y = df['MEDV'].values
X = X.reshape(-1,1)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
quadratic = PolynomialFeatures(degree=3)
X_quad = quadratic.fit_transform(X)
X_quad
```

```python
lr.fit(X_quad,y)
```

```python
ypred = lr.predict(X_quad)
```

```python
plt.scatter(X,y)
plt.plot(X,ypred)
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
plt.show()
```

```python
sorted_X = X_quad[np.argsort(X_quad[:,1])]
ypred = lr.predict(sorted_X)
plt.scatter(X,y)
plt.plot(sorted_X[:,1],ypred)
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
plt.show()
```

```python
mse_quad = mean_squared_error(y,lr.predict(X_quad))
r2_quad = r2_score(y,lr.predict(X_quad))
print("Quadratic regression mse is: %2.2f and R2 is: %2.2f" % (mse_quad,r2_quad))
```

```python
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor(max_depth=3)
tree.fit(X,y)
```

```python
sorted_X = X[np.argsort(X[:,0])]
ypred = tree.predict(sorted_X)
plt.scatter(X,y)
plt.plot(sorted_X,ypred,color='r')
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor(n_estimators=1000,criterion='mse',random_state=1)
```

```python
forest.fit(X,y)
ypred = forest.predict(X)
mse_forest = mean_squared_error(y,ypred)
r2_forest = r2_score(y,ypred)
print("Random Forest mse is: %2.2f and R2 is: %2.2f" % (mse_forest,r2_forest))
```
